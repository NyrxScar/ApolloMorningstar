
<h8><i>üîπ 4. Quais s√£o as orienta√ß√µes sexuais mais comuns?</i></h8>



````python

# Carrega os dados
dados = pd.read_csv('trans-characters-in-youth-cartoons.csv')

# Contar as ocorr√™ncias de cada orienta√ß√£o sexual
contagem_orientacoes = dados['orientation'].value_counts()

# Cria um dicion√°rio para traduzir os termos para portugu√™s
mapeamento_orientacoes = {
    'Gay': 'Gay/Homossexual',
    'Lesbian': 'L√©sbica',
    'Bisexual': 'Bissexual',
    'Queer': 'Queer',
    'Unknown': 'Desconhecida',
    'Undetermined': 'N√£o determinada',
    'Pansexual': 'Pansexual',
    'Asexual, Aromantic': 'Assexual/Arrom√¢ntico',
    'Fluid': 'Fluida',
    'Polysexual': 'Polissexual'
}

# Aqui, renomeamos os √≠ndices da contagem original com os nomes em portugu√™s, ou seja, Aplicar tradu√ß√£o e agrupar termos similares

orientacoes_traduzidas = contagem_orientacoes.rename(index=mapeamento_orientacoes)

# Agrupa as orienta√ß√µes com o mesmo nome traduzido (caso tenha sido mapeado de forma duplicada)
orientacoes_traduzidas = orientacoes_traduzidas.groupby(orientacoes_traduzidas.index).sum()

# Juntar termos similares
orientacoes_traduzidas = orientacoes_traduzidas.groupby(orientacoes_traduzidas.index).sum()

# Ordenar do mais comum para o menos comum
orientacoes_traduzidas = orientacoes_traduzidas.sort_values(ascending=False)

# Mostrar os resultados
orientacoes_traduzidas
````

<h8><i>
üîπ 7. Como a diversidade racial √© representada entre esses personagens?

````python
import pandas as pd

# Carregar os dados
dados = pd.read_csv('trans-characters-in-youth-cartoons.csv')

# Identidades trans (como definido anteriormente)
identidades_trans = [
    'Nonbinary', 'Genderfluid', 'Trans, Woman', 'Trans, Man',
    'Agender', 'Genderless', 'Nonbinary Femme', 'Trans, Nonbinary'
]

# Filtrar personagens trans
trans = dados[dados['gender'].isin(identidades_trans)]

# Limpar e categorizar dados raciais
distribuicao_racial = (
    trans['race'] # Seleciona a coluna Race
    .replace({'POC': 'Pessoa de Cor', 'White': 'Branca'}) # Converte para os termos em pt-br
    .fillna('N√£o especificado')    # Trata valores faltantes/NaN:, Substitui NaN por 'N√£o especificado'
    .value_counts() # Quantifica
    .rename_axis('Identifica√ß√£o Racial') # Renomeia
    .rename('Quantidade') # Renomeia
)

# Mostrar tabela formatada
distribuicao_racial.to_frame()
````

<h3><b>üìå 5. Visualiza√ß√£o/Gr√°ficos </b> </h3>


<h4><b>Gr√°fico de barra de Crescimento de Personagens Trans em Desenhos por Ano </b> </h4>


````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Carrega o arquivo CSV com os dados dos personagens
dados = pd.read_csv('trans-characters-in-youth-cartoons.csv')

# Filtra apenas os personagens que se identificam como trans ou com identidades de g√™nero diversas
trans = dados[dados['gender'].isin([
    'Nonbinary', 'Genderfluid', 'Trans, Woman',
    'Trans, Man', 'Agender', 'Genderless',
    'Nonbinary Femme', 'Trans, Nonbinary'
])]

# Conta quantos personagens trans apareceram por ano e ordena pelo ano
contagem = trans['year'].value_counts().sort_index()

# Cria a √°rea do gr√°fico com tamanho definido
plt.figure(figsize=(12, 6))

# Gera uma lista de cores com gradiente arco-√≠ris (uma cor por barra)
cores = plt.cm.rainbow(np.linspace(0, 1, len(contagem)))

# Plota o gr√°fico de barras com os dados
bars = plt.bar(
    contagem.index.astype(str),  # converte anos para string (para usar no eixo x)
    contagem.values,             # valores (quantidade de personagens)
    color=cores,                 # cores em gradiente
    edgecolor='black'           # borda preta para cada barra
)

# Adiciona os n√∫meros no topo de cada barra (quantidade de personagens)
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,  # posi√ß√£o central da barra
        height,                           # altura do topo
        str(int(height)),                # valor num√©rico (sem casas decimais)
        ha='center',                     # alinhamento horizontal
        va='bottom'                      # alinhamento vertical
    )

# Adiciona t√≠tulo e r√≥tulos nos eixos
plt.title('Personagens Trans em Desenhos por Ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade')

# Rotaciona os r√≥tulos do eixo x para melhor leitura
plt.xticks(rotation=45)

# Adiciona uma grade no eixo y com estilo leve
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Ajusta automaticamente o layout para evitar cortes
plt.tight_layout()

# Exibe o gr√°fico
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Carrega os dados do arquivo CSV
dados = pd.read_csv('trans-characters-in-youth-cartoons.csv')

# Preenche valores ausentes na coluna 'orientation' com 'N√£o especificada'
orientacao = dados['orientation'].fillna('N√£o especificada')

# Traduz os valores da coluna 'orientation' para portugu√™s
orientacao = orientacao.replace({
    'Gay': 'Gay/Homossexual',
    'Lesbian': 'L√©sbica',
    'Bisexual': 'Bissexual',
    'Queer': 'Queer',
    'Unknown': 'Desconhecida',
    'Pansexual': 'Pansexual',
    'Asexual, Aromantic': 'Assexual/Arrom√¢ntico',
    'Bisexual/Pansexual/Polysexual': 'Bi/Pan/Polissexual',
    'Lesbian/Bisexual': 'L√©sbica/Bissexual',
    'Undetermined': 'N√£o determinada'
})

# Conta quantas vezes cada orienta√ß√£o aparece
contagem = orientacao.value_counts()

# Calcula a porcentagem de cada categoria
percentual = (contagem / contagem.sum()) * 100

# Mant√©m apenas categorias com 5.4% ou mais
contagem_modificada = contagem[percentual >= 5.4]

# Soma as categorias com menos de 5.4%
contagem_outros = contagem[percentual < 5.4].sum()

# Adiciona a categoria "Outros" se necess√°rio
if contagem_outros > 0:
    contagem_modificada['Outros'] = contagem_outros

# Cria a √°rea do gr√°fico (figura e eixos)
fig, ax = plt.subplots(figsize=(10, 8))

# Gera uma paleta de cores com at√© 20 varia√ß√µes
cores = plt.cm.tab20.colors

# Cria o gr√°fico de pizza com porcentagens
wedges, texts, autotexts = ax.pie(
    contagem_modificada,               # dados a plotar
    labels=None,                       # sem r√≥tulos diretos
    autopct='%1.1f%%',                 # mostra porcentagem com 1 casa decimal
    startangle=90,                     # come√ßa o gr√°fico do topo
    pctdistance=0.8,                   # dist√¢ncia dos n√∫meros ao centro
    colors=cores[:len(contagem_modificada)],  # cores correspondentes
    wedgeprops={'edgecolor': 'white', 'linewidth': 1}  # bordas brancas finas
)

# Adiciona c√≠rculo branco no centro (forma um gr√°fico de rosca/donut)
centro = plt.Circle((0, 0), 0.70, fc='white')
ax.add_artist(centro)

# Cria uma legenda ao lado do gr√°fico
ax.legend(
    wedges,                            # partes do gr√°fico
    contagem_modificada.index,        # nomes das categorias
    title="Orienta√ß√£o Sexual",        # t√≠tulo da legenda
    loc="center left",                # posi√ß√£o lateral esquerda
    bbox_to_anchor=(1, 0.5)           # deslocamento da legenda
)

# Define o t√≠tulo do gr√°fico
plt.title('Distribui√ß√£o das Sexualidades dos Personagens', y=1.05)

# Ajusta o layout para n√£o cortar partes do gr√°fico
plt.tight_layout()

# Exibe o gr√°fico
plt.show()
````
