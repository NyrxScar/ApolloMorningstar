````python

# Carregar os dados
planilha = pd.read_csv('trans-characters-in-youth-cartoons.csv')

# Renomear colunas
planilha = planilha.rename(columns={
    'character': 'Personagem',
    'show': 'Série',
    'gender': 'Gênero',
    'race': 'Etnia',
    'role': 'Papel',
    'studio': 'Estúdio',
    'representation': 'Representação',
    'orientation': 'Orientação',
    'year': 'Ano',
    'platform': 'Plataforma',
    'confirmation': 'Confirmação',
    'disability': 'Deficiência',
    'rating': 'Classificação'
})

# Tradução dos valores
# Gênero
planilha['Gênero'] = planilha['Gênero'].replace({
    'Cis, Woman': 'Mulher Cis',
    'Cis, Man': 'Homem Cis',
    'Nonbinary': 'Não-binário',
    'Genderfluid': 'Gênero Fluido',
    'Woman': 'Mulher',
    'Man': 'Homem',
    'Agender': 'Agênero',
    'Genderless': 'Sem Gênero',
    'Unknown': 'Não especificado',
    'Intersex': 'Intersexo',
    'Trans, Woman': 'Mulher Trans',
    'Trans, Man': 'Homem Trans',
    'Masc': 'Masculino',
    'Femme': 'Feminino'
})

# Representação
planilha['Representação'] = planilha['Representação'].replace({
    'Explicit': 'Explícita',
    'Implicit': 'Implícita',
    'Undetermined': 'Indeterminada',
    'Suggested': 'Sugerida'
})

# Papel
planilha['Papel'] = planilha['Papel'].replace({
    'Main Character': 'Protagonista',
    'Recurring Character': 'Recorrente',
    'Guest Character': 'Convidado',
    'Supporting Character': 'Coadjuvante',
    'Background Character': 'Figurante'
})

# Etnia
planilha['Etnia'] = planilha['Etnia'].replace({
    'POC': 'Pessoa Não-Branca',
    'White': 'Branca',
    'N/A': 'Não informado',
    'Undetermined': 'Não determinado',
    'Unknown': 'Desconhecido',
    'Asian': 'Asiática',
    'Black': 'Negra',
    'Latino': 'Latina'
})

# Orientação Sexual
planilha['Orientação'] = planilha['Orientação'].replace({
    'Gay': 'Gay',
    'Lesbian': 'Lésbica',
    'Bisexual': 'Bissexual',
    'Queer': 'Queer',
    'Pansexual': 'Pansexual',
    'Asexual': 'Assexual',
    'Aromantic': 'Arromântico',
    'Unknown': 'Não especificada',
    'Undetermined': 'Não determinada',
    'Polysexual': 'Polissexual',
    'Fluid': 'Fluida',
    'Questioning': 'Em questionamento'
})

# Deficiência
planilha['Deficiência'] = planilha['Deficiência'].replace({
    'Autistic': 'Autista',
    'Blurred-Vision': 'Baixa visão',
    'Deaf': 'Surdez',
    'Amputee/Limb-Difference': 'Amputação',
    'ADHD/Asthma': 'TDAH/Asma',
    'PTSD': 'TEPT',
    'Low-Vision': 'Baixa visão',
    None: 'Nenhuma',
    'N/A': 'Não aplicável',
    'Neurodivergent': 'Neurodivergente',
    'Physical Disability': 'Deficiência física',
    'Mental Health Condition': 'Condição de saúde mental'
})

# Confirmação
planilha['Confirmação'] = planilha['Confirmação'].replace({
    'Social Media': 'Redes Sociais',
    'Kiss': 'Beijo',
    'Creative Team Interview': 'Entrevista dos Criadores',
    'Marketing': 'Marketing',
    'Pronouns': 'Pronomes',
    'Identity in Dialogue': 'Identidade no Diálogo',
    'Romantic Partnership': 'Parceria Romântica',
    'Parenting': 'Parentalidade',
    'Companion Material': 'Material Complementar',
    'LGBTQ Symbols/Symbolism': 'Símbolos LGBTQ',
    'Voice Actor': 'Dublador/Dubladora'
})

# Classificação
planilha['Classificação'] = planilha['Classificação'].replace({
    'TV-Y7': 'Livre (acima de 7 anos)',
    'TV-PG': 'Orientação Parental',
    'TV-Y': 'Livre para todos',
    'TV-G': 'Livre (geral)',
    None: 'Não classificada'
})


planilha
````


<h8><i>🔹 4. Quais são as orientações sexuais mais comuns?</i></h8>



````python

# Carrega os dados
dados = pd.read_csv('trans-characters-in-youth-cartoons.csv')

# Contar as ocorrências de cada orientação sexual
contagem_orientacoes = dados['orientation'].value_counts()

# Cria um dicionário para traduzir os termos para português
mapeamento_orientacoes = {
    'Gay': 'Gay/Homossexual',
    'Lesbian': 'Lésbica',
    'Bisexual': 'Bissexual',
    'Queer': 'Queer',
    'Unknown': 'Desconhecida',
    'Undetermined': 'Não determinada',
    'Pansexual': 'Pansexual',
    'Asexual, Aromantic': 'Assexual/Arromântico',
    'Fluid': 'Fluida',
    'Polysexual': 'Polissexual'
}

# Aqui, renomeamos os índices da contagem original com os nomes em português, ou seja, Aplicar tradução e agrupar termos similares

orientacoes_traduzidas = contagem_orientacoes.rename(index=mapeamento_orientacoes)

# Agrupa as orientações com o mesmo nome traduzido (caso tenha sido mapeado de forma duplicada)
orientacoes_traduzidas = orientacoes_traduzidas.groupby(orientacoes_traduzidas.index).sum()

# Juntar termos similares
orientacoes_traduzidas = orientacoes_traduzidas.groupby(orientacoes_traduzidas.index).sum()

# Ordenar do mais comum para o menos comum
orientacoes_traduzidas = orientacoes_traduzidas.sort_values(ascending=False)

# Mostrar os resultados
orientacoes_traduzidas
````

<h8><i>
🔹 7. Como a diversidade racial é representada entre esses personagens?
</i></h8>

````python
import pandas as pd

# Carregar os dados
dados = pd.read_csv('trans-characters-in-youth-cartoons.csv')

# Identidades trans (como definido anteriormente)
identidades_trans = [
    'Nonbinary', 'Genderfluid', 'Trans, Woman', 'Trans, Man',
    'Agender', 'Genderless', 'Nonbinary Femme', 'Trans, Nonbinary'
]

# Filtrar personagens trans
trans = dados[dados['gender'].isin(identidades_trans)]

# Limpar e categorizar dados raciais
distribuicao_racial = (
    trans['race'] # Seleciona a coluna Race
    .replace({'POC': 'Pessoa de Cor', 'White': 'Branca'}) # Converte para os termos em pt-br
    .fillna('Não especificado')    # Trata valores faltantes/NaN:, Substitui NaN por 'Não especificado'
    .value_counts() # Quantifica
    .rename_axis('Identificação Racial') # Renomeia
    .rename('Quantidade') # Renomeia
)

# Mostrar tabela formatada
distribuicao_racial.to_frame()
````

<h8><i>🔹 10. Qual a forma mais comum de confirmação da identidade trans dos personagens?
</i></h8>

````python
import pandas as pd

# Carrega os dados
dados = pd.read_csv('trans-characters-in-youth-cartoons.csv')

# Filtra os personagens trans
personagens_trans = dados[dados['gender'].str.contains('Nonbinary|Genderfluid|Trans|Agender|Genderless', na=False)]

# Conta e renomeia os 3 métodos mais comuns de confirmação da identidade trans:
top_3_metodos = (
    personagens_trans['confirmation']      # Seleciona a coluna com os métodos de confirmação
    .str.split(', ')                       # Divide as strings em listas, separando múltiplos métodos por vírgula
    .explode()                             # Transforma cada item da lista em uma linha separada (1 método por linha)
    .value_counts()                        # Conta quantas vezes cada método aparece
    .head(3)                               # Seleciona os 3 métodos mais frequentes
    .rename_axis('Método de Confirmação')  # Renomeia o índice (os métodos) para uma apresentação mais clara
    .rename('Quantidade')                  # Renomeia a série com o nome 'Quantidade'
)


# Tabela com os resultados
top_3_metodos.to_frame()
````

<h4><b>Gráfico de barra de Crescimento de Personagens Trans em Desenhos por Ano </b> </h4>


````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Carrega o arquivo CSV com os dados dos personagens
dados = pd.read_csv('trans-characters-in-youth-cartoons.csv')

# Filtra apenas os personagens que se identificam como trans ou com identidades de gênero diversas
trans = dados[dados['gender'].isin([
    'Nonbinary', 'Genderfluid', 'Trans, Woman',
    'Trans, Man', 'Agender', 'Genderless',
    'Nonbinary Femme', 'Trans, Nonbinary'
])]

# Conta quantos personagens trans apareceram por ano e ordena pelo ano
contagem = trans['year'].value_counts().sort_index()

# Cria a área do gráfico com tamanho definido
plt.figure(figsize=(12, 6))

# Gera uma lista de cores com gradiente arco-íris (uma cor por barra)
cores = plt.cm.rainbow(np.linspace(0, 1, len(contagem)))

# Plota o gráfico de barras com os dados
bars = plt.bar(
    contagem.index.astype(str),  # converte anos para string (para usar no eixo x)
    contagem.values,             # valores (quantidade de personagens)
    color=cores,                 # cores em gradiente
    edgecolor='black'           # borda preta para cada barra
)

# Adiciona os números no topo de cada barra (quantidade de personagens)
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,  # posição central da barra
        height,                           # altura do topo
        str(int(height)),                # valor numérico (sem casas decimais)
        ha='center',                     # alinhamento horizontal
        va='bottom'                      # alinhamento vertical
    )

# Adiciona título e rótulos nos eixos
plt.title('Personagens Trans em Desenhos por Ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade')

# Rotaciona os rótulos do eixo x para melhor leitura
plt.xticks(rotation=45)

# Adiciona uma grade no eixo y com estilo leve
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Ajusta automaticamente o layout para evitar cortes
plt.tight_layout()

# Exibe o gráfico
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Carrega os dados do arquivo CSV
dados = pd.read_csv('trans-characters-in-youth-cartoons.csv')

# Preenche valores ausentes na coluna 'orientation' com 'Não especificada'
orientacao = dados['orientation'].fillna('Não especificada')

# Traduz os valores da coluna 'orientation' para português
orientacao = orientacao.replace({
    'Gay': 'Gay/Homossexual',
    'Lesbian': 'Lésbica',
    'Bisexual': 'Bissexual',
    'Queer': 'Queer',
    'Unknown': 'Desconhecida',
    'Pansexual': 'Pansexual',
    'Asexual, Aromantic': 'Assexual/Arromântico',
    'Bisexual/Pansexual/Polysexual': 'Bi/Pan/Polissexual',
    'Lesbian/Bisexual': 'Lésbica/Bissexual',
    'Undetermined': 'Não determinada'
})

# Conta quantas vezes cada orientação aparece
contagem = orientacao.value_counts()

# Calcula a porcentagem de cada categoria
percentual = (contagem / contagem.sum()) * 100

# Mantém apenas categorias com 5.4% ou mais
contagem_modificada = contagem[percentual >= 5.4]

# Soma as categorias com menos de 5.4%
contagem_outros = contagem[percentual < 5.4].sum()

# Adiciona a categoria "Outros" se necessário
if contagem_outros > 0:
    contagem_modificada['Outros'] = contagem_outros

# Cria a área do gráfico (figura e eixos)
fig, ax = plt.subplots(figsize=(10, 8))

# Gera uma paleta de cores com até 20 variações
cores = plt.cm.tab20.colors

# Cria o gráfico de pizza com porcentagens
wedges, texts, autotexts = ax.pie(
    contagem_modificada,               # dados a plotar
    labels=None,                       # sem rótulos diretos
    autopct='%1.1f%%',                 # mostra porcentagem com 1 casa decimal
    startangle=90,                     # começa o gráfico do topo
    pctdistance=0.8,                   # distância dos números ao centro
    colors=cores[:len(contagem_modificada)],  # cores correspondentes
    wedgeprops={'edgecolor': 'white', 'linewidth': 1}  # bordas brancas finas
)

# Adiciona círculo branco no centro (forma um gráfico de rosca/donut)
centro = plt.Circle((0, 0), 0.70, fc='white')
ax.add_artist(centro)

# Cria uma legenda ao lado do gráfico
ax.legend(
    wedges,                            # partes do gráfico
    contagem_modificada.index,        # nomes das categorias
    title="Orientação Sexual",        # título da legenda
    loc="center left",                # posição lateral esquerda
    bbox_to_anchor=(1, 0.5)           # deslocamento da legenda
)

# Define o título do gráfico
plt.title('Distribuição das Sexualidades dos Personagens', y=1.05)

# Ajusta o layout para não cortar partes do gráfico
plt.tight_layout()

# Exibe o gráfico
plt.show()
````
